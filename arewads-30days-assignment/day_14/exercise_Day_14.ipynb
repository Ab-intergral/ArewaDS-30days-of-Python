##### Question 3

```python
# Define a call function before map, filter or reduce
from functools import reduce

# Define the call function
def call(func, *args, **kwargs):
    return func(*args, **kwargs)

# Example usage with map
result_map = call(map, lambda x: x * 2, [1, 2, 3])
print(list(result_map))  # Output: [2, 4, 6]

# Example usage with filter
result_filter = call(filter, lambda x: x % 2 == 0, [1, 2, 3, 4, 5])
print(list(result_filter))  # Output: [2, 4]

# Example usage with reduce
result_reduce = call(reduce, lambda x, y: x + y, [1, 2, 3, 4, 5])
print(result_reduce)  # Output: 15 (sum of elements from 1 to 5)
```

```python
# Create a function returning a dictionary, where keys 
# stand for starting letters of countries and values are the number of country names starting with that letter.
def count_countries_starting_with_letter(country_list):
    country_dict = {}

    for country in country_list:
        first_letter = country[0].upper()
        if first_letter in country_dict:
            country_dict[first_letter] += 1
        else:
            country_dict[first_letter] = 1

    return country_dict

starting_letter_counts = count_countries_starting_with_letter(countries)
print(starting_letter_counts)
```
